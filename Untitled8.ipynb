#Justificación y explicación de la normalización#
-Regresión Logística: es un modelo basado en optimización de coeficientes para cada variable. Si una variable tiene escala muy grande (ejemplo: gasto en miles) y otra muy chica (ejemplo: cantidad de meses), el algoritmo puede darle más peso solo por escala, no por información real. La estandarización (media 0, desv estándar 1) iguala esa escala, evitando sesgos y acelerando la convergencia.

-Random Forest: es un modelo basado en árboles que divide por reglas de umbrales. No usa distancias ni optimización por gradiente, por eso no importa si las variables tienen escalas diferentes. Normalizar no aporta nada y puede incluso perder tiempo computacional.

#Evaluación de los Modelos

```python
def evaluar_modelo(modelo, X_test, y_test, nombre_modelo):
    y_pred = modelo.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"--- Evaluación {nombre_modelo} ---")
    print(f"Exactitud (Accuracy): {acc:.4f}")
    print(f"Precisión: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"Matriz de confusión:\n{cm}\n")

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Matriz de confusión - {nombre_modelo}')
    plt.xlabel('Predicción')
    plt.ylabel('Real')
    plt.show()

# Evaluar Regresión Logística
evaluar_modelo(pipeline_lr, X_test, y_test, "Regresión Logística")

# Evaluar Random Forest
evaluar_modelo(pipeline_rf, X_test, y_test, "Random Forest")
```

#Análisis crítico y comparación
Exactitud (Accuracy): mide la proporción de predicciones correctas, pero puede ser engañosa si hay desbalance.

Precisión: de todos los casos que el modelo predijo como cancelación, cuántos realmente cancelaron.

Recall: de todos los que cancelaron, cuántos detectó el modelo (importante para minimizar falsos negativos).

F1-score: balance entre precisión y recall, útil para evaluar en conjunto.

#Sobre overfitting y underfitting
#Overfitting:

Si en entrenamiento el modelo tiene accuracy altísima pero baja en test, o métricas muy superiores en train que test, es señal de overfitting.

Modelos complejos (Random Forest con muchos árboles, profundidad alta) suelen overfitear si no limitás parámetros.

Para corregir: podés limitar profundidad (max_depth), número de árboles, o usar regularización.

#Underfitting:

Si el modelo tiene métricas bajas tanto en entrenamiento como en test, no aprende patrones bien.

Puede pasar con modelos muy simples (por ejemplo, regresión logística sin variables significativas).

Para corregir: agregar más variables, usar modelos más complejos, o ajustar parámetros.
